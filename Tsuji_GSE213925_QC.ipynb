```python
import numpy as np
import pandas as pd
import scanpy as sc
import anndata
```

```python
sc.settings.verbosity = 3
sc.logging.print_header()
sc.settings.set_figure_params(dpi=80, facecolor='white')
```

```python
results_file = 'write/Tsuji_GSE213925_QC.h5ad'
```

```python
adata_wtNSG01 = sc.read_10x_mtx(
    'GSE213925/GSE213925_RAW/GSM6596509_wtNSG01',  
    cache=True)  
```

```python
adata_wtNSG01
```

```python
adata_wtNSG01.obs
```

```python
adata_wtNSG01.var
```

```python
sc.pl.highest_expr_genes(adata_wtNSG01, n_top=20)
```

```python
adata_wtNSG01.var['mt'] = adata_wtNSG01.var_names.str.startswith('mt-') 
```

```python
adata_wtNSG01.var['Hbb'] = adata_wtNSG01.var_names.str.startswith('Hbb-') 
```

```python
sc.pp.calculate_qc_metrics(adata_wtNSG01, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
```

```python
sc.pp.calculate_qc_metrics(adata_wtNSG01, qc_vars=['Hbb'], percent_top=None, log1p=False, inplace=True)
```

```python
sc.pl.violin(adata_wtNSG01, ['n_genes_by_counts', 'total_counts'], jitter=0.4, rotation= 45)
sc.pl.violin(adata_wtNSG01, ['pct_counts_mt'], jitter=0.4, rotation= 45)
sc.pl.violin(adata_wtNSG01, ['pct_counts_Hbb'], jitter=0.4, rotation= 45)
```

```python
adata_wtNSG01 = adata_wtNSG01[adata_wtNSG01.obs['pct_counts_mt'] < 10, :]
```

```python
adata_wtNSG01 = adata_wtNSG01[adata_wtNSG01.obs['pct_counts_mt'] < 20, :]
```

```python
adata_wtNSG01
```

```python
sc.pp.filter_cells(adata_wtNSG01, min_genes=200)
sc.pp.filter_genes(adata_wtNSG01, min_cells=3)
```

```python
adata_wtNSG01
```

```python
adata_wtNSG01.raw = adata_wtNSG01
```

```python
adata_wtNSG01.write(results_file)
```
